# 16. Spark 기반 웨이퍼 데이터 파이프라인

**WM-811K → Medallion Architecture (Bronze / Silver / Gold)**

| 기술 | 역할 |
|------|------|
| PySpark DataFrame API | 분산 데이터 처리 |
| Spark SQL | Hive 호환 SQL 분석 |
| Window 함수 | SPC 이동 평균/Cpk |
| Parquet + Partitioning | 컬럼형 저장, 파티션 푸시다운 |
| UDF | 웨이퍼 맵 바이너리 → 공간 특징 |

```
LSWMD.pkl / all_maps_resized.npy
         ↓
   [Bronze Parquet]
         ↓ JOIN + 공간통계 UDF
   [Silver Parquet]  ← partitioned by failure_type
         ↓
   [Gold Parquet]    ← train / val / test + Feature Store + SPC
```

## 2. Silver 계층 — JOIN + 공간 통계

- `wafer_id` 기준 LEFT JOIN (맵 + 공정 파라미터)
- `failure_type` 파티셔닝 (Hive 파티션 푸시다운 최적화)
- UDF: 맵 바이너리 → `defect_density`, `edge_defect_ratio`, `center_defect_ratio`

````python
from spark_pipeline.transform.bronze_to_silver import bronze_to_silver

t0 = time.time()
silver = bronze_to_silver(
    spark,
    bronze_maps_path   = PATHS['bronze_maps'],
    bronze_params_path = PATHS['bronze_params'],
    out_path           = PATHS['silver_labeled'],
)
print(f'소요 시간: {time.time()-t0:.1f}s')
````

````python
# Spark SQL: 결함 유형별 공정 파라미터 평균
silver_df = spark.read.parquet(PATHS['silver_labeled'])
silver_df.createOrReplaceTempView('silver')

spark.sql("""
    SELECT
        failure_type,
        COUNT(*)                    AS n,
        ROUND(AVG(cmp_pressure),2)  AS avg_cmp,
        ROUND(AVG(particle_count),2)AS avg_particles,
        ROUND(AVG(defect_density),4)AS avg_defect_density,
        ROUND(AVG(annealing_temp),1)AS avg_anneal_temp
    FROM silver
    GROUP BY failure_type
    ORDER BY n DESC
""").show(truncate=False)
````

## 4. Feature Store — 공간 통계 특징 추출 (UDF)

PySpark UDF로 웨이퍼 맵(4096 bytes) → 7차원 벡터.
`wafer_map_flat` 없이 경량 Feature Store 생성.

````python
from spark_pipeline.feature_engineering.spatial_features import build_feature_store

feats = build_feature_store(
    spark,
    silver_path = PATHS['silver_labeled'],
    out_path    = PATHS['gold_features'],
)
````

````python
# Spark SQL: 결함 유형별 공간 특징 비교
fs = spark.read.parquet(PATHS['gold_features'])
fs.createOrReplaceTempView('features')

spark.sql("""
    SELECT
        failure_type,
        ROUND(AVG(defect_density),4)      AS avg_density,
        ROUND(AVG(edge_defect_ratio),3)   AS avg_edge_ratio,
        ROUND(AVG(center_defect_ratio),3) AS avg_center_ratio,
        ROUND(AVG(radial_mean),2)         AS avg_radial_mean
    FROM features
    WHERE failure_type != 'none'
    GROUP BY failure_type
    ORDER BY avg_density DESC
""").show(truncate=False)
````

## 5. SPC 집계 — Spark SQL Window 함수

- 이동 평균 (25매 롤링 윈도우)
- Cpk = MIN((UCL-μ)/3σ, (μ-LCL)/3σ)
- OOC 경보 뷰

````python
from spark_pipeline.feature_engineering.spc_aggregation import build_spc_statistics

spc, summary = build_spc_statistics(
    spark,
    silver_path = PATHS['silver_labeled'],
    out_path    = PATHS['gold_spc'],
)
````

````python
# SPC 이동 평균 시각화 (CMP 압력)
import matplotlib.pyplot as plt
import pandas as pd

spc_df = spark.read.parquet(PATHS['gold_spc'])
spc_pd = (spc_df
    .select('wafer_id', 'failure_type', 'cmp_pressure', 'cmp_pressure_roll25')
    .filter('failure_type IN ("none", "Edge-Ring", "Near-full")')
    .orderBy('wafer_id')
    .limit(500)
    .toPandas()
)

fig, ax = plt.subplots(figsize=(12, 4))
for ft, grp in spc_pd.groupby('failure_type'):
    ax.scatter(grp['wafer_id'], grp['cmp_pressure'], s=8, alpha=0.5, label=ft)
ax.plot(spc_pd['wafer_id'].values, spc_pd['cmp_pressure_roll25'].values,
        'k-', lw=1.5, label='Roll-25 Avg')
ax.axhline(120, color='red',    ls='--', lw=1, label='UCL=120')
ax.axhline(80,  color='orange', ls='--', lw=1, label='LCL=80')
ax.set_xlabel('Wafer ID')
ax.set_ylabel('CMP Pressure (psi)')
ax.set_title('CMP Pressure SPC Chart (500 samples)')
ax.legend(loc='upper right', fontsize=8)
plt.tight_layout()
plt.savefig('analysis/spc_cmp_pressure.png', dpi=120)
plt.show()
print('SPC 차트 저장: analysis/spc_cmp_pressure.png')
````

## 6. Hive 테이블 등록 (Spark SQL Catalog)

`enableHiveSupport()` 없이도 Spark SQL 카탈로그에 임시 테이블 등록 가능.
프로덕션 환경에서는 Hive Metastore 연결 후 영구 등록.

````python
# Parquet 경로를 Spark SQL 임시 뷰로 등록
for name, path in [
    ('bronze_maps',   PATHS['bronze_maps']),
    ('silver',        PATHS['silver_labeled']),
    ('gold_train',    PATHS['gold_train']),
    ('gold_features', PATHS['gold_features']),
    ('gold_spc',      PATHS['gold_spc']),
]:
    spark.read.parquet(path).createOrReplaceTempView(name)
    print(f'  뷰 등록: {name}')

print('\n등록된 뷰 목록:')
spark.sql('SHOW VIEWS').show()
````

````python
# Hive QL 스타일 분석 쿼리
spark.sql("""
    SELECT
        s.failure_type,
        COUNT(*) AS n,
        ROUND(AVG(s.cmp_pressure), 2)   AS avg_cmp,
        ROUND(AVG(f.defect_density), 4) AS avg_density,
        ROUND(AVG(f.edge_defect_ratio), 3) AS avg_edge_ratio,
        ROUND(STDDEV(s.cmp_pressure), 2) AS std_cmp
    FROM silver s
    JOIN gold_features f ON s.wafer_id = f.wafer_id
    WHERE s.failure_type != 'none'
    GROUP BY s.failure_type
    ORDER BY avg_density DESC
""").show(truncate=False)
````

## 7. 데이터 레이크 저장 현황 요약

````python
from pathlib import Path

lake_root = PROJECT_ROOT / 'data' / 'lake'
print('데이터 레이크 저장 현황')
print('=' * 55)
total = 0
for layer in ['bronze', 'silver', 'gold']:
    ld = lake_root / layer
    if not ld.exists():
        continue
    parquets = list(ld.rglob('*.parquet'))
    size = sum(p.stat().st_size for p in parquets)
    total += size
    print(f'  {layer.upper():<8}: {len(parquets):>3} files, {size/1024/1024:>7.1f} MB')
print(f'  {'TOTAL':<8}: {total/1024/1024:>7.1f} MB')

# Spark 통계
print('\nGold 분할 통계:')
train_n = spark.read.parquet(PATHS['gold_train']).count()
val_n   = spark.read.parquet(PATHS['gold_val']).count()
test_n  = spark.read.parquet(PATHS['gold_test']).count()
print(f'  Train: {train_n:,} | Val: {val_n:,} | Test: {test_n:,}')
````
